```python
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
from sklearn import preprocessing
```

```python
data_unif = 12*np.random.random(1000) +4
data_gauss = np.random.normal(-3,4,1000)
data_exp = np.random.exponential(4,1000)
data_unif_small = np.random.random(1000) -1
data_gauss_small = np.random.normal(8,0.3,1000)
data_exp_small = np.random.exponential(1,1000)
data_gauss_ok = np.random.normal(0,1,1000)

df = pd.DataFrame(dict(uniform = data_unif, gauss=data_gauss, exp=data_exp, gauss_ok = data_gauss_ok,
                      uniform_s = data_unif_small, gauss_s=data_gauss_small, exp_s=data_exp_small))
```

```python
for i,k in enumerate(df.columns):
    if i!=3:
        continue
    color='.7'
    if 'ok' in k:
        color = 'C2'
    plot = sns.displot(df[k], height=2, aspect=1, color=color)
    plot.axes[0,0].set_xlabel('Data')
```

```python
df_ = df.stack().reset_index()
df_.columns = ['sample','distribution','data']
df_['ok'] = ['ok' if k=='gauss_ok' else 'not' for k in df_.distribution]
sns.displot(df_, x='data', col='distribution', bins=np.arange(-5,11,.5), height=2, col_wrap=4, hue='ok')
```

```python
df.plot.hist(bins=50, alpha=.3)
```

```python
fig, ax = plt.subplots(figsize=(3,3))
sns.scatterplot(df, x='gauss', y='gauss_s', ax=ax)
ax.axis('equal')
ax.grid()
ax.set_ylabel('Feature 2')
ax.set_xlabel('Feature 1')

X = df[['gauss','gauss_s']]

scaler = preprocessing.StandardScaler().fit(X)
X_transform = scaler.transform(X)

fig, ax = plt.subplots(figsize=(3,3))
ax.scatter(X_transform[:,0], X_transform[:,1], zorder=100)
ax.axis('equal')
ax.grid()
ax.set_ylabel('Standardized Feature 2')
ax.set_xlabel('Standardized Feature 1')
```

```python
fig, ax = plt.subplots(figsize=(3,3))
sns.scatterplot(df, x='uniform', y='uniform_s', ax=ax)
ax.axis('equal')
ax.grid()
ax.set_ylabel('Feature 2')
ax.set_xlabel('Feature 1')

X = df[['uniform','uniform_s']]

scaler = preprocessing.MinMaxScaler().fit(X)
X_transform = scaler.transform(X)
fig, ax = plt.subplots(figsize=(3,3))
ax.scatter(X_transform[:,0], X_transform[:,1], zorder=100)
ax.axis('equal')
ax.grid()
ax.set_ylabel('Standardized Feature 2')
ax.set_xlabel('Standardized Feature 1')
```

```python
fig, ax = plt.subplots(figsize=(3,3))
sns.scatterplot(df, x='exp', y='exp_s', ax=ax)
# ax.axis('equal')
ax.grid()
ax.set_ylabel('Feature 2')
ax.set_xlabel('Feature 1')

X = df[['exp','exp_s']]

scaler = preprocessing.PowerTransformer(method='box-cox').fit(X)
X_transform = scaler.transform(X)

fig, ax = plt.subplots(figsize=(3,3))
ax.scatter(X_transform[:,0], X_transform[:,1], zorder=100)
ax.axis('equal')
ax.grid()
ax.set_ylabel('Transformed Feature 2')
ax.set_xlabel('Transformed Feature 1')
```
