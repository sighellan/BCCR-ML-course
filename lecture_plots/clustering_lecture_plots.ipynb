```python
import sklearn
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.concat([np.random.normal(5,1,1000),np.random.normal(-1,2.5,1000), np.random.normal(5,1.7,1000), np.random.normal(-5,0.5,1000) ])
y = np.concat([np.random.normal(-5,1,1000),np.random.normal(2,2.5,1000), np.random.normal(9,1,1000) , np.random.normal(10,1,1000)])
X = np.stack([x,y]).T
```

```python
fig, ax = plt.subplots(figsize=(4,4,))
ax.scatter(x,y, zorder=10, color='.4', s=5)
ax.set_xlim(-10,10)
ax.set_ylim(-10,15)
ax.set_xlabel('Feature 1')
ax.set_ylabel('Feature 2')
ax.grid(zorder=0, color='.4')
plt.tight_layout()
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=4)
model.fit(X)
```

```python
labels = model.predict(X)
```

```python
x_min, x_max = -10,10
y_min, y_max = -10, 15
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 300),
                     np.linspace(y_min, y_max, 300))
```

```python

fig, ax = plt.subplots(figsize=(4,4,))
ax.scatter(x,y, zorder=10, color='.4', s=10, edgecolor='.9', linewidth=.4)
ax.set_xlim(-10,10)
ax.set_ylim(-10,15)
ax.set_xlabel('Feature 1')
ax.set_ylabel('Feature 2')
ax.grid(zorder=0, color='.4')
Z = model.predict(np.stack([xx.ravel(), yy.ravel()]).T)
Z = Z.reshape(xx.shape)
ax.pcolormesh(xx,yy,Z, cmap='tab20', alpha=.5)
plt.tight_layout()
```
